```python
import matplotlib.pyplot as plt
```

```python
my_colors = ['#5cb85c','#5bc0de','#d9534f']
df_ds_sorted = df_ds.sort_values(by='Very interested', ascending=False)

perc = (df_ds_sorted /2233) *100 
perc = round (perc, 2)
perc

perc.plot(kind='bar',
           figsize=(20,8),
           width=0.9,
           color= my_colors)


plt.title('Percentage of respondents Interest in Data Science Areas ',fontsize=16)
plt.ylabel('Percentage',fontsize=14)
plt.xlabel('Data Science Area',fontsize=14)

```

```python
colors_list = ['#5cb85c','#5bc0de','#d9534f']

results = perc.div(perc.sum(1), axis=0)

ax = results.plot(kind='bar',figsize=(20,8),width = 0.8,color = colors_list,edgecolor=None)
plt.legend(labels=perc.columns,fontsize= 14)
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 16)

plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center')
```
